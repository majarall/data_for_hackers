```python
# Discrete distributions 
from random import choice, choices, sample, shuffle

outcomes = ["win", "lose", "draw", "play agian", "double win"]

choice(outcomes)
```

```python
choices(outcomes, k=10)
```

```python
Counter(choices(outcomes, k=10))
```

```python
# Give each category a weight
# Its basically saying that given some outcomes with biased weightings, what is the count for each out come when spinning the wheel 10000
Counter(choices(outcomes,[5, 4, 3, 2, 1], k=10000))
```

```python
outcomes
```

```python
# works inplace
shuffle(outcomes)
```

```python
sample(outcomes, k=4)
```

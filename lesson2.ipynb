```python
# Statitical significance and p-values
# P-values: Means what is the chance that what we observed is due to chance and real effect

# Statistical significance of the difference of two means

drug = [7.1, 8.5, 6.4, 7.7, 8.2, 7.6, 8.4, 5.1, 8.1, 7.4, 6.9, 8.4]
placebo = [8.2, 6.1, 7.1, 4.9, 7.4, 8.1, 7.1, 6.2, 7.0, 6.6, 6.3]
```

```python
mean(drug)
```

```python
mean(placebo)
```

```python
obs_diff = mean(drug) - mean(placebo)

combination = drug + placebo
number_drugs = len(drug)

shuffle(combination)
```

```python
# If we reshuffle (permuting, relabeling) he particpant
# Is the new mean diff the same or more extreme thanthe observed


def trial():
    shuffle(combination)
    drug = combination[:number_drugs]
    placebo = combination[number_drugs:]
    new_diff = mean(drug) - mean(placebo)
    return new_diff >= obs_diff

n = 10000
sum(trial() for i in range(n))/n       # This is called the p-value Likelihood that the outcome was solely due to chance 




```

```python
# Single server queue simulation 

average_arrival_interval = 5.6
average_service_time = 5.0
stdev_service_time = 0.5

num_waiting = 0
arrivals = []
starts = []
arrival = service_end = 0.0
n = 20_000
for i in range(n):
    if arrival <= service_end:
        num_waiting += 1
        arrival += expovariate(1.0 / average_arrival_interval)
        arrivals.append(arrival)

    else:
        num_waiting -= 1
        service_start = service_end if num_waiting else arrival
        service_time = gauss(average_service_time, stdev_service_time)
        service_end = service_start + service_time 
        starts.append(service_start)

waits = [start - arrival for arrival, start in zip(arrivals, starts)]
print(f"Mean wait {mean(waits):.2f} Stdev wait: {stdev(waits):.2f}")
print(f"Median wait {median(waits):.2f} Max wait: {max(waits):.2f}")

```
